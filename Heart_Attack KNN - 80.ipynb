```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Harsh/Heart Attack/heart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
df.drop(['sex'],axis = 1,inplace = True)
```

```python
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X=sc.fit_transform(X)
#X_test=sc.fit_transform(X_test)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X,y)
```

```python
df['output'].value_counts()
```

```python
# f-1 Score is used for find the acuracy of target classes{ 1 }, reffered to imbalance data

from sklearn.metrics import f1_score
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)


print(f1_score(y_train,pred_train))
print(f1_score(y_test,pred_test))

print(f1_score(y_train,pred_train,average=None))
print(f1_score(y_test,pred_test,average=None))
```

```python
# Cross Validation Score
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(model,X,y,cv= 5)
cvs
```

```python
# --> Grid search
from sklearn.model_selection import GridSearchCV
gsv = GridSearchCV(model,param_grid={'n_neighbors':[6,7,8]},cv = 3)
gsv.fit(X,y)
print(gsv.best_score_)
print(gsv.best_params_)
```

```python
# --> Randomized search
from sklearn.model_selection import RandomizedSearchCV
rsv = RandomizedSearchCV(model,param_distributions={'n_neighbors':[2,3,4,5,6]},cv = 4, n_iter= 5)
rsv.fit(X,y)
print(rsv.best_score_)
print(rsv.best_params_)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
